1. Se desea estimar la altura promedio de estudiantes universitarios. En una muestra de 50 estudiantes,
se obtuvo una media de 170 cm. Se sabe que la desviación estándar poblacional es de 10 cm.
Construye un intervalo de confianza del 95 % para la media poblacional.



```python
import numpy as np
from scipy import stats

# Datos proporcionados
n = 50  # Tamaño de la muestra
media_muestral = 170  # Media de la muestra en cm
desviacion_estandar_poblacional = 10  # Desviación estándar poblacional en cm
nivel_confianza = 0.95  # Nivel de confianza deseado

# Calcular el error estándar de la media
# Fórmula: Error Estándar = Desviación Estándar Poblacional / sqrt(n)
error_estandar = desviacion_estandar_poblacional / np.sqrt(n)

# Calcular el valor crítico de Z para el nivel de confianza del 95%
# Para un intervalo de confianza bilateral del 95%, el área en cada cola es (1 - 0.95) / 2 = 0.025
# Usamos la función ppf (percent point function) de la distribución normal estándar para encontrar el valor Z
valor_z = stats.norm.ppf(1 - (1 - nivel_confianza) / 2)

# Calcular el margen de error
# Fórmula: Margen de Error = Valor Z * Error Estándar
margen_error = valor_z * error_estandar

# Calcular el intervalo de confianza
# Fórmula: Intervalo de Confianza = [Media Muestral - Margen de Error, Media Muestral + Margen de Error]
intervalo_inferior = media_muestral - margen_error
intervalo_superior = media_muestral + margen_error

# Imprimir los resultados
print(f"Tamaño de la muestra (n): {n}")
print(f"Media de la muestra: {media_muestral} cm")
print(f"Desviación estándar poblacional: {desviacion_estandar_poblacional} cm")
print(f"Nivel de confianza: {nivel_confianza}")
print(f"Error estándar de la media: {error_estandar:.2f} cm")
print(f"Valor crítico de Z para el {nivel_confianza*100}% de confianza: {valor_z:.2f}")
print(f"Margen de error: {margen_error:.2f} cm")
print(f"Intervalo de confianza del {nivel_confianza*100}% para la media poblacional: [{intervalo_inferior:.2f} cm, {intervalo_superior:.2f} cm]")

# Opcional: Graficar la distribución normal y el intervalo de confianza
import matplotlib.pyplot as plt

# Rango de valores para graficar la distribución normal
x = np.linspace(media_muestral - 4 * error_estandar, media_muestral + 4 * error_estandar, 100)
# Función de densidad de probabilidad (PDF) de la distribución normal
pdf = stats.norm.pdf(x, loc=media_muestral, scale=error_estandar)

plt.figure(figsize=(10, 6))
plt.plot(x, pdf, label='Distribución muestral de la media')

# Marcar la media muestral
plt.axvline(media_muestral, color='red', linestyle='dashed', linewidth=1, label=f'Media muestral ({media_muestral} cm)')

# Marcar el intervalo de confianza
plt.fill_between(x, 0, pdf, where=(x >= intervalo_inferior) & (x <= intervalo_superior), color='skyblue', alpha=0.5, label=f'Intervalo de confianza del {nivel_confianza*100}%')

# Configurar el gráfico
plt.title('Distribución muestral de la media y Intervalo de Confianza')
plt.xlabel('Altura (cm)')
plt.ylabel('Densidad de probabilidad')
plt.legend()
plt.grid(True)
plt.show()
```

6. Una cadena de cafeterías desea estimar la temperatura promedio (°C) del café que sirve a sus clientes. Se tomó una muestra aleatoria de 12 tazas donde la media fue de 84°C, y la desviación estándar
5°C. Con un nivel de confianza del 95 % calcule el intervalo de confianza para la temperatura media
poblacional del café.

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Datos proporcionados para la temperatura del café
n_cafe = 12  # Tamaño de la muestra de tazas de café
media_muestral_cafe = 84  # Media de la muestra de temperatura en °C
desviacion_estandar_muestral_cafe = 5  # Desviación estándar *muestral* de temperatura en °C
nivel_confianza_cafe = 0.95  # Nivel de confianza deseado

# Calcular los grados de libertad
grados_libertad_cafe = n_cafe - 1

# Calcular el error estándar de la media (usando la desviación estándar muestral)
# Fórmula: Error Estándar = Desviación Estándar Muestral / sqrt(n)
error_estandar_cafe = desviacion_estandar_muestral_cafe / np.sqrt(n_cafe)

# Calcular el valor crítico de t para el nivel de confianza del 95% y los grados de libertad
# Para un intervalo de confianza bilateral del 95%, el área en cada cola es (1 - 0.95) / 2 = 0.025
# Usamos la función ppf (percent point function) de la distribución t de Student para encontrar el valor t
valor_t_cafe = stats.t.ppf(1 - (1 - nivel_confianza_cafe) / 2, grados_libertad_cafe)

# Calcular el margen de error
# Fórmula: Margen de Error = Valor t * Error Estándar
margen_error_cafe = valor_t_cafe * error_estandar_cafe

# Calcular el intervalo de confianza
# Fórmula: Intervalo de Confianza = [Media Muestral - Margen de Error, Media Muestral + Margen de Error]
intervalo_inferior_cafe = media_muestral_cafe - margen_error_cafe
intervalo_superior_cafe = media_muestral_cafe + margen_error_cafe

# Imprimir los resultados para la temperatura del café
print(f"Tamaño de la muestra (n): {n_cafe}")
print(f"Media de la muestra: {media_muestral_cafe} °C")
print(f"Desviación estándar muestral: {desviacion_estandar_muestral_cafe} °C")
print(f"Nivel de confianza: {nivel_confianza_cafe}")
print(f"Grados de libertad: {grados_libertad_cafe}")
print(f"Error estándar de la media (usando desviación estándar muestral): {error_estandar_cafe:.2f} °C")
print(f"Valor crítico de t para el {nivel_confianza_cafe*100}% de confianza: {valor_t_cafe:.2f}")
print(f"Margen de error: {margen_error_cafe:.2f} °C")
print(f"Intervalo de confianza del {nivel_confianza_cafe*100}% para la media poblacional: [{intervalo_inferior_cafe:.2f} °C, {intervalo_superior_cafe:.2f} °C]")

# Graficar la distribución t de Student y el intervalo de confianza
# Rango de valores para graficar la distribución t
x_cafe = np.linspace(media_muestral_cafe - 4 * error_estandar_cafe, media_muestral_cafe + 4 * error_estandar_cafe, 100)
# Función de densidad de probabilidad (PDF) de la distribución t de Student
pdf_cafe = stats.t.pdf(x_cafe, df=grados_libertad_cafe, loc=media_muestral_cafe, scale=error_estandar_cafe)

plt.figure(figsize=(10, 6))
plt.plot(x_cafe, pdf_cafe, label=f'Distribución t ({grados_libertad_cafe} df)')

# Marcar la media muestral
plt.axvline(media_muestral_cafe, color='red', linestyle='dashed', linewidth=1, label=f'Media muestral ({media_muestral_cafe} °C)')

# Marcar el intervalo de confianza
plt.fill_between(x_cafe, 0, pdf_cafe, where=(x_cafe >= intervalo_inferior_cafe) & (x_cafe <= intervalo_superior_cafe), color='skyblue', alpha=0.5, label=f'Intervalo de confianza del {nivel_confianza_cafe*100}%')

# Configurar el gráfico
plt.title('Distribución t de Student y Intervalo de Confianza para la Temperatura del Café')
plt.xlabel('Temperatura (°C)')
plt.ylabel('Densidad de probabilidad')
plt.legend()
plt.grid(True)
plt.show()
```

11. El departamento de salud planea una encuesta para estimar la proporción de residentes vacunados
contra la influenza. En un estudio piloto, el 72 % de una muestra mostró estar vacunado. Determine
el tamaño de muestra necesario para estimar esta proporción con un margen de error del 3 % y un
nivel de confianza del 95 %. Si no se tuviera el estudio piloto, ¿cuál sería el tamaño requerido?


```python
import numpy as np
from scipy import stats

# Datos proporcionados
margen_error = 0.03  # Margen de error deseado (3%)
nivel_confianza = 0.95  # Nivel de confianza deseado (95%)
proporcion_piloto = 0.72  # Proporción de vacunados en el estudio piloto (72%)

# Calcular el valor crítico de Z para el nivel de confianza del 95%
# Para un intervalo de confianza bilateral del 95%, el área en cada cola es (1 - 0.95) / 2 = 0.025
valor_z = stats.norm.ppf(1 - (1 - nivel_confianza) / 2)

# --- Caso 1: Usando la proporción del estudio piloto ---
# Fórmula para el tamaño de muestra (con estimación previa de la proporción):
# n = (Z^2 * p * (1-p)) / ME^2
# Donde:
# Z = Valor crítico de Z
# p = Proporción estimada (del estudio piloto)
# ME = Margen de error

n_con_piloto = (valor_z**2 * proporcion_piloto * (1 - proporcion_piloto)) / (margen_error**2)

# Redondear hacia arriba para asegurar el margen de error deseado
n_con_piloto = np.ceil(n_con_piloto)

print("--- Tamaño de muestra con estudio piloto ---")
print(f"Margen de error deseado: {margen_error*100}%")
print(f"Nivel de confianza: {nivel_confianza*100}%")
print(f"Proporción estimada (estudio piloto): {proporcion_piloto:.2f}")
print(f"Valor crítico de Z: {valor_z:.2f}")
print(f"Tamaño de muestra calculado: {int(n_con_piloto)}")
print("-" * 30)

# --- Caso 2: Sin un estudio piloto (usando la proporción más conservadora, p=0.5) ---
# Cuando no se tiene una estimación previa de la proporción, se utiliza p = 0.5.
# Esto maximiza el producto p*(1-p) y, por lo tanto, el tamaño de muestra,
# asegurando que el margen de error no sea mayor al deseado, independientemente de la proporción real.
proporcion_conservadora = 0.5

# Fórmula para el tamaño de muestra (sin estimación previa):
# n = (Z^2 * 0.25) / ME^2  (ya que p*(1-p) es máximo en p=0.5, dando 0.25)

n_sin_piloto = (valor_z**2 * proporcion_conservadora * (1 - proporcion_conservadora)) / (margen_error**2)

# Redondear hacia arriba
n_sin_piloto = np.ceil(n_sin_piloto)

print("--- Tamaño de muestra sin estudio piloto ---")
print(f"Margen de error deseado: {margen_error*100}%")
print(f"Nivel de confianza: {nivel_confianza*100}%")
print(f"Proporción utilizada (conservadora): {proporcion_conservadora:.2f}")
print(f"Valor crítico de Z: {valor_z:.2f}")
print(f"Tamaño de muestra calculado: {int(n_sin_piloto)}")
print("-" * 30)
```

```python
import numpy as np
from scipy import stats

# Datos proporcionados para la encuesta política
margen_error_politica = 0.02  # Margen de error deseado (2%)
nivel_confianza_politica = 0.99  # Nivel de confianza deseado (99%)
proporcion_previa_politica = 0.42  # Proporción de apoyo en la encuesta previa (42%)

# Calcular el valor crítico de Z para el nivel de confianza del 99%
# Para un intervalo de confianza bilateral del 99%, el área en cada cola es (1 - 0.99) / 2 = 0.005
valor_z_politica = stats.norm.ppf(1 - (1 - nivel_confianza_politica) / 2)

# --- Caso 1: Usando la proporción de la encuesta previa ---
# Fórmula para el tamaño de muestra (con estimación previa de la proporción):
# n = (Z^2 * p * (1-p)) / ME^2
# Donde:
# Z = Valor crítico de Z
# p = Proporción estimada (de la encuesta previa)
# ME = Margen de error

n_con_previa_politica = (valor_z_politica**2 * proporcion_previa_politica * (1 - proporcion_previa_politica)) / (margen_error_politica**2)

# Redondear hacia arriba para asegurar el margen de error deseado
n_con_previa_politica = np.ceil(n_con_previa_politica)

print("--- Tamaño de muestra con encuesta previa ---")
print(f"Margen de error deseado: {margen_error_politica*100}%")
print(f"Nivel de confianza: {nivel_confianza_politica*100}%")
print(f"Proporción estimada (encuesta previa): {proporcion_previa_politica:.2f}")
print(f"Valor crítico de Z: {valor_z_politica:.2f}")
print(f"Tamaño de muestra calculado: {int(n_con_previa_politica)}")
print("-" * 30)

# --- Caso 2: Sin una encuesta previa (usando la proporción más conservadora, p=0.5) ---
# Cuando no se tiene una estimación previa de la proporción, se utiliza p = 0.5.
# Esto maximiza el producto p*(1-p) y, por lo tanto, el tamaño de muestra,
# asegurando que el margen de error no sea mayor al deseado, independientemente de la proporción real.
proporcion_conservadora_politica = 0.5

# Fórmula para el tamaño de muestra (sin estimación previa):
# n = (Z^2 * 0.25) / ME^2  (ya que p*(1-p) es máximo en p=0.5, dando 0.25)

n_sin_previa_politica = (valor_z_politica**2 * proporcion_conservadora_politica * (1 - proporcion_conservadora_politica)) / (margen_error_politica**2)

# Redondear hacia arriba
n_sin_previa_politica = np.ceil(n_sin_previa_politica)

print("--- Tamaño de muestra sin encuesta previa ---")
print(f"Margen de error deseado: {margen_error_politica*100}%")
print(f"Nivel de confianza: {nivel_confianza_politica*100}%")
print(f"Proporción utilizada (conservadora): {proporcion_conservadora_politica:.2f}")
print(f"Valor crítico de Z: {valor_z_politica:.2f}")
print(f"Tamaño de muestra calculado: {int(n_sin_previa_politica)}")
print("-" * 30)
```

**Justificación para usar p = 0.5 cuando no hay datos preliminares:**

La fórmula para calcular el tamaño de muestra para una proporción es:

$n = \frac{Z^2 * p * (1-p)}{ME^2}$

Donde:
*   $n$ es el tamaño de muestra necesario.
*   $Z$ es el valor crítico de Z para el nivel de confianza deseado.
*   $p$ es la proporción estimada de la población.
*   $ME$ es el margen de error deseado.

El término $p * (1-p)$ en el numerador de la fórmula representa la variabilidad de la proporción. Queremos encontrar el valor de $p$ que maximiza este producto, ya que un mayor valor de $p * (1-p)$ resultará en un tamaño de muestra más grande. Esto es importante porque, si usamos un tamaño de muestra basado en el valor máximo de la variabilidad, podemos estar seguros de que el margen de error real será igual o menor que el margen de error deseado, independientemente de cuál sea la verdadera proporción poblacional.

Consideremos algunos valores de $p * (1-p)$:
*   Si $p = 0.1$, $p * (1-p) = 0.1 * 0.9 = 0.09$
*   Si $p = 0.3$, $p * (1-p) = 0.3 * 0.7 = 0.21$
*   Si $p = 0.5$, $p * (1-p) = 0.5 * 0.5 = 0.25$
*   Si $p = 0.7$, $p * (1-p) = 0.7 * 0.3 = 0.21$
*   Si $p = 0.9$, $p * (1-p) = 0.9 * 0.1 = 0.09$

Como se puede observar, el producto $p * (1-p)$ alcanza su valor máximo de 0.25 cuando $p = 0.5$.

Por lo tanto, cuando no se tiene una estimación previa de la proporción poblacional ($p$), se utiliza $p = 0.5$ en la fórmula del tamaño de muestra. Esto se hace para obtener el **tamaño de muestra más conservador (más grande)** posible. Al usar este tamaño de muestra, se garantiza que el margen de error de la estimación será, como máximo, el margen de error deseado, sin importar cuál sea la verdadera proporción en la población. En esencia, es una forma de ser precavido y asegurar que la muestra sea lo suficientemente grande para la estimación, incluso en el peor de los casos de variabilidad.

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Datos proporcionados
media_poblacional_hipotetizada = 500  # μ0: Peso promedio afirmado por la empresa en gramos
n_muestra = 40  # Tamaño de la muestra
media_muestral = 495  # Media de la muestra en gramos
desviacion_estandar_poblacional = 15  # σ: Desviación estándar poblacional en gramos
nivel_significancia = 0.05  # α: Nivel de significancia

# 1. Plantear las hipótesis nula y alternativa
# H0: La máquina está calibrada correctamente (μ = 500 g)
# H1: La máquina está subllenando (μ < 500 g) - Prueba unilateral izquierda

print("1. Plantear las hipótesis nula y alternativa:")
print(f"   H0: μ = {media_poblacional_hipotetizada} g (El peso promedio es 500 g)")
print(f"   H1: μ < {media_poblacional_hipotetizada} g (El peso promedio es menor a 500 g - Prueba unilateral izquierda)")
print("-" * 30)

# 2. Calcular el estadístico de prueba (Estadístico Z)
# Fórmula: Z = (media_muestral - μ0) / (σ / sqrt(n))
error_estandar = desviacion_estandar_poblacional / np.sqrt(n_muestra)
estadistico_z = (media_muestral - media_poblacional_hipotetizada) / error_estandar

print("2. Calcule el estadístico Z:")
print(f"   Error estándar de la media: {error_estandar:.2f} g")
print(f"   Estadístico Z calculado: {estadistico_z:.2f}")
print("-" * 30)

# 3. Determinar el valor crítico y la región de rechazo
# Para una prueba unilateral izquierda con α = 0.05, el valor crítico de Z es el valor
# tal que el área a su izquierda en la distribución normal estándar es α.
valor_critico_z = stats.norm.ppf(nivel_significancia)

print("3. Determine el valor crítico y la región de rechazo:")
print(f"   Nivel de significancia (α): {nivel_significancia}")
print(f"   Valor crítico de Z (prueba unilateral izquierda): {valor_critico_z:.2f}")
print(f"   Región de rechazo: Z < {valor_critico_z:.2f}")
print("-" * 30)

# 4. Calcular el valor p
# Para una prueba unilateral izquierda, el valor p es el área a la izquierda del estadístico Z calculado
valor_p = stats.norm.cdf(estadistico_z)

print("5. Calcule el estadístico de prueba y el valor p:") # Re-imprimir estadístico Z for clarity
print(f"   Estadístico Z calculado: {estadistico_z:.2f}")
print(f"   Valor p: {valor_p:.4f}")
print("-" * 30)


# 6. Graficar la distribución normal estándar, región de rechazo, valor crítico y estadístico calculado
plt.figure(figsize=(10, 6))
x = np.linspace(-4, 4, 100)
plt.plot(x, stats.norm.pdf(x, 0, 1), label='Distribución Normal Estándar')

# Región de rechazo (unilateral izquierda)
x_rechazo = np.linspace(-4, valor_critico_z, 100)
plt.fill_between(x_rechazo, 0, stats.norm.pdf(x_rechazo, 0, 1), color='red', alpha=0.5, label='Región de Rechazo (α = 0.05)')

# Valor crítico
plt.axvline(valor_critico_z, color='red', linestyle='dashed', linewidth=1, label=f'Valor Crítico Z ({valor_critico_z:.2f})')

# Estadístico Z calculado
plt.axvline(estadistico_z, color='blue', linestyle='dashed', linewidth=1, label=f'Estadístico Z Calculado ({estadistico_z:.2f})')

# Configurar el gráfico
plt.title('Prueba de Hipótesis Unilateral Izquierda (Z-test)')
plt.xlabel('Valor Z')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()


# 7. Tomar una decisión estadística y concluir
print("7. Tome una decisión estadística y concluya:")
print(f"   Nivel de significancia (α): {nivel_significancia}")
print(f"   Valor p: {valor_p:.4f}")

if valor_p < nivel_significancia:
    decision = "Rechazar H0"
    conclusion = f"Con un nivel de significancia de {nivel_significancia}, hay suficiente evidencia estadística para rechazar la hipótesis nula (H0). Esto sugiere que el peso promedio de los paquetes de cereal es significativamente menor a {media_poblacional_hipotetizada} g. Podemos concluir que la máquina de llenado probablemente está descalibrada y subllenando los paquetes."
else:
    decision = "No rechazar H0"
    conclusion = f"Con un nivel de significancia de {nivel_significancia}, no hay suficiente evidencia estadística para rechazar la hipótesis nula (H0). Esto sugiere que la diferencia observada en la media muestral ({media_muestral} g) respecto a la media hipotetizada ({media_poblacional_hipotetizada} g) podría deberse simplemente a la variación aleatoria del muestreo. No podemos concluir que la máquina esté subllenando basándonos en esta muestra."

print(f"   Decisión: {decision}")
print(f"   Conclusión: {conclusion}")
print("\nExplicación del valor p:")
print(f"   El valor p ({valor_p:.4f}) representa la probabilidad de obtener una media muestral tan extrema o más extrema ({media_muestral} g) que la observada, asumiendo que la hipótesis nula (H0) es verdadera (es decir, que el peso promedio real es 500 g).")
print(f"   Dado que el valor p ({valor_p:.4f}) es {'menor' if valor_p < nivel_significancia else 'mayor o igual'} que el nivel de significancia (α = {nivel_significancia}), {'rechazamos' if valor_p < nivel_significancia else 'no rechazamos'} H0.")
```

23. Un profesor afirma que el promedio de su clase en un examen nacional es diferente al promedio
nacional de 75 puntos. Una muestra de 36 estudiantes de su clase tiene una media de 78. La
desviación estándar poblacional es σ = 12. Con un nivel de significancia: α = 0,05.


```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Datos proporcionados
media_poblacional_hipotetizada = 75  # μ0: Promedio nacional del examen
n_muestra = 36  # Tamaño de la muestra de estudiantes de la clase
media_muestral = 78  # Media de la muestra de la clase
desviacion_estandar_poblacional = 12  # σ: Desviación estándar poblacional del examen
nivel_significancia = 0.05  # α: Nivel de significancia

# 1. Plantear las hipótesis nula y alternativa
# H0: El promedio de la clase es igual al promedio nacional (μ = 75 puntos)
# H1: El promedio de la clase es diferente al promedio nacional (μ ≠ 75 puntos) - Prueba bilateral

print("1. Plantear las hipótesis nula y alternativa:")
print(f"   H0: μ = {media_poblacional_hipotetizada} puntos (El promedio de la clase es igual al promedio nacional)")
print(f"   H1: μ ≠ {media_poblacional_hipotetizada} puntos (El promedio de la clase es diferente al promedio nacional - Prueba bilateral)")
print("-" * 30)

# 2. Calcular el estadístico de prueba (Estadístico Z)
# Fórmula: Z = (media_muestral - μ0) / (σ / sqrt(n))
error_estandar = desviacion_estandar_poblacional / np.sqrt(n_muestra)
estadistico_z = (media_muestral - media_poblacional_hipotetizada) / error_estandar

print("2. Calcule el estadístico Z:")
print(f"   Error estándar de la media: {error_estandar:.2f} puntos")
print(f"   Estadístico Z calculado: {estadistico_z:.2f}")
print("-" * 30)

# 3. Determinar los valores críticos y la región de rechazo
# Para una prueba bilateral con α = 0.05, el área en cada cola es α / 2 = 0.025.
# Los valores críticos de Z son los valores que dejan 0.025 en cada cola.
valor_critico_z_inferior = stats.norm.ppf(nivel_significancia / 2)
valor_critico_z_superior = stats.norm.ppf(1 - nivel_significancia / 2)

print("3. Determine los valores críticos y la región de rechazo:")
print(f"   Nivel de significancia (α): {nivel_significancia}")
print(f"   Valores críticos de Z (prueba bilateral): {valor_critico_z_inferior:.2f} y {valor_critico_z_superior:.2f}")
print(f"   Región de rechazo: Z < {valor_critico_z_inferior:.2f} o Z > {valor_critico_z_superior:.2f}")
print("-" * 30)

# 4. Calcular el valor p
# Para una prueba bilateral, el valor p es el doble del área en la cola correspondiente al estadístico Z calculado.
# Si Z es negativo, el valor p es 2 * P(Z <= estadistico_z). Si Z es positivo, el valor p es 2 * P(Z >= estadistico_z).
# Equivalentemente, el valor p es 2 * el área en la cola más extrema.
valor_p = 2 * (1 - stats.norm.cdf(abs(estadistico_z))) # Usamos abs() para obtener el área en la cola correcta y multiplicamos por 2

print("5. Calcule el estadístico de prueba y el valor p:") # Re-imprimir estadístico Z for clarity
print(f"   Estadístico Z calculado: {estadistico_z:.2f}")
print(f"   Valor p: {valor_p:.4f}")
print("-" * 30)


# 6. Graficar la distribución normal estándar, región de rechazo, valores críticos y estadístico calculado
plt.figure(figsize=(10, 6))
x = np.linspace(-4, 4, 100)
plt.plot(x, stats.norm.pdf(x, 0, 1), label='Distribución Normal Estándar')

# Región de rechazo (bilateral)
x_rechazo_inferior = np.linspace(-4, valor_critico_z_inferior, 100)
plt.fill_between(x_rechazo_inferior, 0, stats.norm.pdf(x_rechazo_inferior, 0, 1), color='red', alpha=0.5, label=f'Región de Rechazo (α/2 = {nivel_significancia/2})')

x_rechazo_superior = np.linspace(valor_critico_z_superior, 4, 100)
plt.fill_between(x_rechazo_superior, 0, stats.norm.pdf(x_rechazo_superior, 0, 1), color='red', alpha=0.5) # No need for label again

# Valores críticos
plt.axvline(valor_critico_z_inferior, color='red', linestyle='dashed', linewidth=1, label=f'Valores Críticos Z ({valor_critico_z_inferior:.2f}, {valor_critico_z_superior:.2f})')
plt.axvline(valor_critico_z_superior, color='red', linestyle='dashed', linewidth=1) # No need for label again

# Estadístico Z calculado
plt.axvline(estadistico_z, color='blue', linestyle='dashed', linewidth=1, label=f'Estadístico Z Calculado ({estadistico_z:.2f})')

# Configurar el gráfico
plt.title('Prueba de Hipótesis Bilateral (Z-test)')
plt.xlabel('Valor Z')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()


# 7. Tomar una decisión estadística y concluir
print("7. Tome una decisión estadística y concluya:")
print(f"   Nivel de significancia (α): {nivel_significancia}")
print(f"   Valor p: {valor_p:.4f}")

if valor_p < nivel_significancia:
    decision = "Rechazar H0"
    conclusion = f"Con un nivel de significancia de {nivel_significancia}, hay suficiente evidencia estadística para rechazar la hipótesis nula (H0). Esto sugiere que el promedio de la clase del profesor ({media_muestral} puntos) es significativamente diferente del promedio nacional ({media_poblacional_hipotetizada} puntos)."
else:
    decision = "No rechazar H0"
    conclusion = f"Con un nivel de significancia de {nivel_significancia}, no hay suficiente evidencia estadística para rechazar la hipótesis nula (H0). Esto sugiere que la diferencia observada en la media muestral ({media_muestral} puntos) respecto al promedio nacional hipotetizado ({media_poblacional_hipotetizada} puntos) podría deberse simplemente a la variación aleatoria del muestreo. No podemos concluir que el promedio de la clase sea significativamente diferente del promedio nacional basándonos en esta muestra."

print(f"   Decisión: {decision}")
print(f"   Conclusión: {conclusion}")
print("\nExplicación del valor p:")
print(f"   El valor p ({valor_p:.4f}) representa la probabilidad de obtener una media muestral tan extrema o más extrema (en cualquier dirección, ya que es una prueba bilateral) como la observada ({media_muestral} puntos), asumiendo que la hipótesis nula (H0) es verdadera (es decir, que el promedio real de la clase es 75 puntos).")
print(f"   Dado que el valor p ({valor_p:.4f}) es {'menor' if valor_p < nivel_significancia else 'mayor o igual'} que el nivel de significancia (α = {nivel_significancia}), {'rechazamos' if valor_p < nivel_significancia else 'no rechazamos'} H0.")
```

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Datos proporcionados
media_poblacional_hipotetizada = 10  # μ0: Reducción promedio de presión arterial afirmada por el laboratorio en mmHg
n_muestra = 25  # Tamaño de la muestra
media_muestral = 8  # Media de la muestra de reducción de presión arterial en mmHg
desviacion_estandar_poblacional = 3  # σ: Desviación estándar poblacional de la reducción en mmHg
nivel_significancia = 0.05  # α: Nivel de significancia

# 1. Plantear las hipótesis nula y alternativa
# H0: El medicamento reduce la presión arterial en 10 mmHg en promedio (μ = 10 mmHg)
# H1: El medicamento no reduce la presión arterial en 10 mmHg en promedio (μ ≠ 10 mmHg) - Prueba bilateral (assuming 'diferente' from previous examples, or could be unilateral if suspecting less reduction)
# Based on the wording "afirma que... reduce en 10" and the sample showing 8, a two-tailed test seems appropriate to see if it's *different* from 10. If the suspicion was specifically that it reduces *less* than 10, it would be unilateral left. Let's assume bilateral for now, as "diferente" is more common unless specified otherwise.

print("1. Plantear las hipótesis nula y alternativa:")
print(f"   H0: μ = {media_poblacional_hipotetizada} mmHg (El medicamento reduce la presión arterial en 10 mmHg en promedio)")
print(f"   H1: μ ≠ {media_poblacional_hipotetizada} mmHg (El medicamento reduce la presión arterial en promedio de forma diferente a 10 mmHg - Prueba bilateral)")
print("-" * 30)

# 2. Calcular el estadístico de prueba (Estadístico Z)
# Fórmula: Z = (media_muestral - μ0) / (σ / sqrt(n))
error_estandar = desviacion_estandar_poblacional / np.sqrt(n_muestra)
estadistico_z = (media_muestral - media_poblacional_hipotetizada) / error_estandar

print("2. Calcule el estadístico Z:")
print(f"   Error estándar de la media: {error_estandar:.2f} mmHg")
print(f"   Estadístico Z calculado: {estadistico_z:.2f}")
print("-" * 30)

# 3. Determinar los valores críticos y la región de rechazo
# Para una prueba bilateral con α = 0.05, el área en cada cola es α / 2 = 0.025.
# Los valores críticos de Z son los valores que dejan 0.025 en cada cola.
valor_critico_z_inferior = stats.norm.ppf(nivel_significancia / 2)
valor_critico_z_superior = stats.norm.ppf(1 - nivel_significancia / 2)

print("3. Determine los valores críticos y la región de rechazo:")
print(f"   Nivel de significancia (α): {nivel_significancia}")
print(f"   Valores críticos de Z (prueba bilateral): {valor_critico_z_inferior:.2f} y {valor_critico_z_superior:.2f}")
print(f"   Región de rechazo: Z < {valor_critico_z_inferior:.2f} o Z > {valor_critico_z_superior:.2f}")
print("-" * 30)

# 4. Calcular el valor p
# Para una prueba bilateral, el valor p es el doble del área en la cola correspondiente al estadístico Z calculado.
valor_p = 2 * (1 - stats.norm.cdf(abs(estadistico_z))) # Usamos abs() para obtener el área en la cola correcta y multiplicamos por 2

print("5. Calcule el estadístico de prueba y el valor p:") # Re-imprimir estadístico Z for clarity
print(f"   Estadístico Z calculado: {estadistico_z:.2f}")
print(f"   Valor p: {valor_p:.4f}")
print("-" * 30)


# 6. Graficar la distribución normal estándar, región de rechazo, valores críticos y estadístico calculado
plt.figure(figsize=(10, 6))
x = np.linspace(-4, 4, 100)
plt.plot(x, stats.norm.pdf(x, 0, 1), label='Distribución Normal Estándar')

# Región de rechazo (bilateral)
x_rechazo_inferior = np.linspace(-4, valor_critico_z_inferior, 100)
plt.fill_between(x_rechazo_inferior, 0, stats.norm.pdf(x_rechazo_inferior, 0, 1), color='red', alpha=0.5, label=f'Región de Rechazo (α/2 = {nivel_significancia/2})')

x_rechazo_superior = np.linspace(valor_critico_z_superior, 4, 100)
plt.fill_between(x_rechazo_superior, 0, stats.norm.pdf(x_rechazo_superior, 0, 1), color='red', alpha=0.5) # No need for label again

# Valores críticos
plt.axvline(valor_critico_z_inferior, color='red', linestyle='dashed', linewidth=1, label=f'Valores Críticos Z ({valor_critico_z_inferior:.2f}, {valor_critico_z_superior:.2f})')
plt.axvline(valor_critico_z_superior, color='red', linestyle='dashed', linewidth=1) # No need for label again

# Estadístico Z calculado
plt.axvline(estadistico_z, color='blue', linestyle='dashed', linewidth=1, label=f'Estadístico Z Calculado ({estadistico_z:.2f})')

# Configurar el gráfico
plt.title('Prueba de Hipótesis Bilateral (Z-test)')
plt.xlabel('Valor Z')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()


# 7. Tomar una decisión estadística y concluir
print("7. Tome una decisión estadística y concluya:")
print(f"   Nivel de significancia (α): {nivel_significancia}")
print(f"   Valor p: {valor_p:.4f}")

if valor_p < nivel_significancia:
    decision = "Rechazar H0"
    conclusion = f"Con un nivel de significancia de {nivel_significancia}, hay suficiente evidencia estadística para rechazar la hipótesis nula (H0). Esto sugiere que la reducción promedio de la presión arterial con este medicamento es significativamente diferente de {media_poblacional_hipotetizada} mmHg."
else:
    decision = "No rechazar H0"
    conclusion = f"Con un nivel de significancia de {nivel_significancia}, no hay suficiente evidencia estadística para rechazar la hipótesis nula (H0). Esto sugiere que la diferencia observada en la media muestral ({media_muestral} mmHg) respecto a la reducción promedio hipotetizada ({media_poblacional_hipotetizada} mmHg) podría deberse simplemente a la variación aleatoria del muestreo. No podemos concluir que la reducción promedio sea significativamente diferente de 10 mmHg basándonos en esta muestra."

print(f"   Decisión: {decision}")
print(f"   Conclusión: {conclusion}")
print("\nExplicación del valor p:")
print(f"   El valor p ({valor_p:.4f}) representa la probabilidad de obtener una media muestral tan extrema o más extrema (en cualquier dirección, ya que es una prueba bilateral) como la observada ({media_muestral} mmHg), asumiendo que la hipótesis nula (H0) es verdadera (es decir, que la reducción promedio real es 10 mmHg).")
print(f"   Dado que el valor p ({valor_p:.4f}) es {'menor' if valor_p < nivel_significancia else 'mayor o igual'} que el nivel de significancia (α = {nivel_significancia}), {'rechazamos' if valor_p < nivel_significancia else 'no rechazamos'} H0.")
```

26. Un restaurante afirma que su tiempo promedio de entrega a domicilio es 25 minutos. Un cliente
sospecha que es mayor. Se registran 15 entregas, obteniendo una media de 28 minutos y desviación
estándar muestral s = 4 minutos. Con un nivel de significancia de α = 0,05.


```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Datos proporcionados
media_poblacional_hipotetizada = 25  # μ0: Tiempo promedio de entrega afirmado por el restaurante en minutos
n_muestra = 15  # Tamaño de la muestra
media_muestral = 28  # Media de la muestra en minutos
desviacion_estandar_muestral = 4  # s: Desviación estándar *muestral* en minutos
nivel_significancia = 0.05  # α: Nivel de significancia

# 1. Plantear las hipótesis nula y alternativa
# H0: El tiempo promedio de entrega es de 25 minutos (μ = 25 minutos)
# H1: El tiempo promedio de entrega es mayor a 25 minutos (μ > 25 minutos) - Prueba unilateral derecha

print("1. Plantear las hipótesis nula y alternativa:")
print(f"   H0: μ = {media_poblacional_hipotetizada} minutos (El tiempo promedio de entrega es 25 minutos)")
print(f"   H1: μ > {media_poblacional_hipotetizada} minutos (El tiempo promedio de entrega es mayor a 25 minutos - Prueba unilateral derecha)")
print("-" * 30)

# 2. Calcular los grados de libertad
grados_libertad = n_muestra - 1

# Calcular el estadístico de prueba (Estadístico t)
# Fórmula: t = (media_muestral - μ0) / (s / sqrt(n))
error_estandar_muestral = desviacion_estandar_muestral / np.sqrt(n_muestra)
estadistico_t = (media_muestral - media_poblacional_hipotetizada) / error_estandar_muestral

print("2. Calcule el estadístico t:")
print(f"   Grados de libertad: {grados_libertad}")
print(f"   Error estándar de la media (usando desviación estándar muestral): {error_estandar_muestral:.2f} minutos")
print(f"   Estadístico t calculado: {estadistico_t:.2f}")
print("-" * 30)

# 3. Determinar el valor crítico y la región de rechazo
# Para una prueba unilateral derecha con α = 0.05 y {grados_libertad} grados de libertad,
# el valor crítico de t es el valor tal que el área a su derecha en la distribución t de Student es α.
valor_critico_t = stats.t.ppf(1 - nivel_significancia, grados_libertad)

print("3. Determine el valor crítico y la región de rechazo:")
print(f"   Nivel de significancia (α): {nivel_significancia}")
print(f"   Valor crítico de t (prueba unilateral derecha, {grados_libertad} df): {valor_critico_t:.2f}")
print(f"   Región de rechazo: t > {valor_critico_t:.2f}")
print("-" * 30)

# 4. Calcular el valor p
# Para una prueba unilateral derecha, el valor p es el área a la derecha del estadístico t calculado
valor_p = 1 - stats.t.cdf(estadistico_t, grados_libertad)

print("5. Calcule el estadístico de prueba y el valor p:") # Re-imprimir estadístico t for clarity
print(f"   Estadístico t calculado: {estadistico_t:.2f}")
print(f"   Valor p: {valor_p:.4f}")
print("-" * 30)


# 6. Graficar la distribución t de Student, región de rechazo, valor crítico y estadístico calculado
plt.figure(figsize=(10, 6))
# Rango de valores para graficar la distribución t (centrada en 0, escalada a 1 para el gráfico estándar)
x_t = np.linspace(-4, 4, 100)
plt.plot(x_t, stats.t.pdf(x_t, grados_libertad), label=f'Distribución t ({grados_libertad} df)')

# Región de rechazo (unilateral derecha)
x_rechazo = np.linspace(valor_critico_t, 4, 100)
plt.fill_between(x_rechazo, 0, stats.t.pdf(x_rechazo, grados_libertad), color='red', alpha=0.5, label=f'Región de Rechazo (α = {nivel_significancia})')

# Valor crítico
plt.axvline(valor_critico_t, color='red', linestyle='dashed', linewidth=1, label=f'Valor Crítico t ({valor_critico_t:.2f})')

# Estadístico t calculado
plt.axvline(estadistico_t, color='blue', linestyle='dashed', linewidth=1, label=f'Estadístico t Calculado ({estadistico_t:.2f})')

# Configurar el gráfico
plt.title('Prueba de Hipótesis Unilateral Derecha (t-test)')
plt.xlabel('Valor t')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()


# 7. Tomar una decisión estadística y concluir
print("7. Tome una decisión estadística y concluya:")
print(f"   Nivel de significancia (α): {nivel_significancia}")
print(f"   Valor p: {valor_p:.4f}")

if valor_p < nivel_significancia:
    decision = "Rechazar H0"
    conclusion = f"Con un nivel de significancia de {nivel_significancia}, hay suficiente evidencia estadística para rechazar la hipótesis nula (H0). Esto sugiere que el tiempo promedio de entrega a domicilio del restaurante ({media_muestral} minutos) es significativamente mayor a {media_poblacional_hipotetizada} minutos."
else:
    decision = "No rechazar H0"
    conclusion = f"Con un nivel de significancia de {nivel_significancia}, no hay suficiente evidencia estadística para rechazar la hipótesis nula (H0). Esto sugiere que la diferencia observada en el tiempo promedio de entrega de la muestra ({media_muestral} minutos) respecto al tiempo hipotetizado ({media_poblacional_hipotetizada} minutos) podría deberse simplemente a la variación aleatoria del muestreo. No podemos concluir que el tiempo promedio de entrega sea significativamente mayor a 25 minutos basándonos en esta muestra."

print(f"   Decisión: {decision}")
print(f"   Conclusión: {conclusion}")
print("\nExplicación del valor p:")
print(f"   El valor p ({valor_p:.4f}) representa la probabilidad de obtener una media muestral tan extrema o más extrema ({media_muestral} minutos) que la observada, asumiendo que la hipótesis nula (H0) es verdadera (es decir, que el tiempo promedio real de entrega es 25 minutos).")
print(f"   Dado que el valor p ({valor_p:.4f}) es {'menor' if valor_p < nivel_significancia else 'mayor o igual'} que el nivel de significancia (α = {nivel_significancia}), {'rechazamos' if valor_p < nivel_significancia else 'no rechazamos'} H0.")
```

estudiantes tiene una media de 75 y s = 8. Con un nivel de significancia de α = 0,01.


```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Datos proporcionados
media_poblacional_hipotetizada = 80  # μ0: Promedio afirmado por el colegio
n_muestra = 12  # Tamaño de la muestra
media_muestral = 75  # Media de la muestra
desviacion_estandar_muestral = 8  # s: Desviación estándar *muestral*
nivel_significancia = 0.01  # α: Nivel de significancia

# 1. Plantear las hipótesis nula y alternativa
# H0: El promedio de la clase es igual al promedio afirmado (μ = 80)
# H1: El promedio de la clase es diferente al promedio afirmado (μ ≠ 80) - Prueba bilateral

print("1. Plantear las hipótesis nula y alternativa:")
print(f"   H0: μ = {media_poblacional_hipotetizada} (El promedio de la clase es igual al promedio afirmado)")
print(f"   H1: μ ≠ {media_poblacional_hipotetizada} (El promedio de la clase es diferente al promedio afirmado - Prueba bilateral)")
print("-" * 30)

# 2. Calcular los grados de libertad
grados_libertad = n_muestra - 1

# Calcular el estadístico de prueba (Estadístico t)
# Fórmula: t = (media_muestral - μ0) / (s / sqrt(n))
error_estandar_muestral = desviacion_estandar_muestral / np.sqrt(n_muestra)
estadistico_t = (media_muestral - media_poblacional_hipotetizada) / error_estandar_muestral

print("2. Calcule el estadístico t:")
print(f"   Grados de libertad: {grados_libertad}")
print(f"   Error estándar de la media (usando desviación estándar muestral): {error_estandar_muestral:.2f}")
print(f"   Estadístico t calculado: {estadistico_t:.2f}")
print("-" * 30)

# 3. Determinar los valores críticos y la región de rechazo
# Para una prueba bilateral con α = 0.01 y {grados_libertad} grados de libertad,
# el área en cada cola es α / 2 = 0.005.
# Los valores críticos de t son los valores que dejan 0.005 en cada cola.
valor_critico_t_inferior = stats.t.ppf(nivel_significancia / 2, grados_libertad)
valor_critico_t_superior = stats.t.ppf(1 - nivel_significancia / 2, grados_libertad)

print("3. Determine los valores críticos y la región de rechazo:")
print(f"   Nivel de significancia (α): {nivel_significancia}")
print(f"   Valores críticos de t (prueba bilateral, {grados_libertad} df): {valor_critico_t_inferior:.2f} y {valor_critico_t_superior:.2f}")
print(f"   Región de rechazo: t < {valor_critico_t_inferior:.2f} o t > {valor_critico_t_superior:.2f}")
print("-" * 30)

# 4. Calcular el valor p
# Para una prueba bilateral, el valor p es el doble del área en la cola correspondiente al estadístico t calculado.
valor_p = 2 * (1 - stats.t.cdf(abs(estadistico_t), grados_libertad)) # Usamos abs() para obtener el área en la cola correcta y multiplicamos por 2

print("5. Calcule el estadístico de prueba y el valor p:") # Re-imprimir estadístico t for clarity
print(f"   Estadístico t calculado: {estadistico_t:.2f}")
print(f"   Valor p: {valor_p:.4f}")
print("-" * 30)


# 6. Graficar la distribución t de Student, región de rechazo, valores críticos y estadístico calculado
plt.figure(figsize=(10, 6))
# Rango de valores para graficar la distribución t (centrada en 0, escalada a 1 para el gráfico estándar)
x_t = np.linspace(-4, 4, 100)
plt.plot(x_t, stats.t.pdf(x_t, grados_libertad), label=f'Distribución t ({grados_libertad} df)')

# Región de rechazo (bilateral)
x_rechazo_inferior = np.linspace(-4, valor_critico_t_inferior, 100)
plt.fill_between(x_rechazo_inferior, 0, stats.t.pdf(x_rechazo_inferior, grados_libertad), color='red', alpha=0.5, label=f'Región de Rechazo (α/2 = {nivel_significancia/2})')

x_rechazo_superior = np.linspace(valor_critico_t_superior, 4, 100)
plt.fill_between(x_rechazo_superior, 0, stats.t.pdf(x_rechazo_superior, grados_libertad), color='red', alpha=0.5) # No need for label again

# Valores críticos
plt.axvline(valor_critico_t_inferior, color='red', linestyle='dashed', linewidth=1, label=f'Valores Críticos t ({valor_critico_t_inferior:.2f}, {valor_critico_t_superior:.2f})')
plt.axvline(valor_critico_t_superior, color='red', linestyle='dashed', linewidth=1) # No need for label again

# Estadístico t calculado
plt.axvline(estadistico_t, color='blue', linestyle='dashed', linewidth=1, label=f'Estadístico t Calculado ({estadistico_t:.2f})')

# Configurar el gráfico
plt.title('Prueba de Hipótesis Bilateral (t-test)')
plt.xlabel('Valor t')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()


# 7. Tomar una decisión estadística y concluir
print("7. Tome una decisión estadística y concluya:")
print(f"   Nivel de significancia (α): {nivel_significancia}")
print(f"   Valor p: {valor_p:.4f}")

if valor_p < nivel_significancia:
    decision = "Rechazar H0"
    conclusion = f"Con un nivel de significancia de {nivel_significancia}, hay suficiente evidencia estadística para rechazar la hipótesis nula (H0). Esto sugiere que el promedio de la clase en matemáticas ({media_muestral}) es significativamente diferente del promedio afirmado por el colegio ({media_poblacional_hipotetizada})."
else:
    decision = "No rechazar H0"
    conclusion = f"Con un nivel de significancia de {nivel_significancia}, no hay suficiente evidencia estadística para rechazar la hipótesis nula (H0). Esto sugiere que la diferencia observada en la media muestral ({media_muestral}) respecto al promedio hipotetizado ({media_poblacional_hipotetizada}) podría deberse simplemente a la variación aleatoria del muestreo. No podemos concluir que el promedio de la clase sea significativamente diferente de 80 basándonos en esta muestra."

print(f"   Decisión: {decision}")
print(f"   Conclusión: {conclusion}")
print("\nExplicación del valor p:")
print(f"   El valor p ({valor_p:.4f}) representa la probabilidad de obtener una media muestral tan extrema o más extrema (en cualquier dirección, ya que es una prueba bilateral) como la observada ({media_muestral}), asumiendo que la hipótesis nula (H0) es verdadera (es decir, que el promedio real de la clase es 80).")
print(f"   Dado que el valor p ({valor_p:.4f}) es {'menor' if valor_p < nivel_significancia else 'mayor o igual'} que el nivel de significancia (α = {nivel_significancia}), {'rechazamos' if valor_p < nivel_significancia else 'no rechazamos'} H0.")
```

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Datos proporcionados
media_poblacional_hipotetizada = 10000  # μ0: Duración promedio afirmada por el fabricante en horas
n_muestra = 20  # Tamaño de la muestra
media_muestral = 9800  # Media de la muestra en horas
desviacion_estandar_muestral = 300  # s: Desviación estándar *muestral* en horas
nivel_significancia = 0.05  # α: Nivel de significancia

# 1. Plantear las hipótesis nula y alternativa
# H0: La duración promedio de los focos es de 10,000 horas (μ = 10000 horas)
# H1: La duración promedio de los focos es diferente de 10,000 horas (μ ≠ 10000 horas) - Prueba bilateral

print("1. Plantear las hipótesis nula y alternativa:")
print(f"   H0: μ = {media_poblacional_hipotetizada} horas (La duración promedio es 10,000 horas)")
print(f"   H1: μ ≠ {media_poblacional_hipotetizada} horas (La duración promedio es diferente de 10,000 horas - Prueba bilateral)")
print("-" * 30)

# 2. Calcular los grados de libertad
grados_libertad = n_muestra - 1

# Calcular el estadístico de prueba (Estadístico t)
# Fórmula: t = (media_muestral - μ0) / (s / sqrt(n))
error_estandar_muestral = desviacion_estandar_muestral / np.sqrt(n_muestra)
estadistico_t = (media_muestral - media_poblacional_hipotetizada) / error_estandar_muestral

print("2. Calcule el estadístico t:")
print(f"   Grados de libertad: {grados_libertad}")
print(f"   Error estándar de la media (usando desviación estándar muestral): {error_estandar_muestral:.2f} horas")
print(f"   Estadístico t calculado: {estadistico_t:.2f}")
print("-" * 30)

# 3. Determinar los valores críticos y la región de rechazo
# Para una prueba bilateral con α = 0.05 y {grados_libertad} grados de libertad,
# el área en cada cola es α / 2 = 0.025.
# Los valores críticos de t son los valores que dejan 0.025 en cada cola.
valor_critico_t_inferior = stats.t.ppf(nivel_significancia / 2, grados_libertad)
valor_critico_t_superior = stats.t.ppf(1 - nivel_significancia / 2, grados_libertad)

print("3. Determine los valores críticos y la región de rechazo:")
print(f"   Nivel de significancia (α): {nivel_significancia}")
print(f"   Valores críticos de t (prueba bilateral, {grados_libertad} df): {valor_critico_t_inferior:.2f} y {valor_critico_t_superior:.2f}")
print(f"   Región de rechazo: t < {valor_critico_t_inferior:.2f} o t > {valor_critico_t_superior:.2f}")
print("-" * 30)

# 4. Calcular el valor p
# Para una prueba bilateral, el valor p es el doble del área en la cola correspondiente al estadístico t calculado.
valor_p = 2 * (1 - stats.t.cdf(abs(estadistico_t), grados_libertad)) # Usamos abs() para obtener el área en la cola correcta y multiplicamos por 2

print("5. Calcule el estadístico de prueba y el valor p:") # Re-imprimir estadístico t for clarity
print(f"   Estadístico t calculado: {estadistico_t:.2f}")
print(f"   Valor p: {valor_p:.4f}")
print("-" * 30)


# 6. Graficar la distribución t de Student, región de rechazo, valores críticos y estadístico calculado
plt.figure(figsize=(10, 6))
# Rango de valores para graficar la distribución t (centrada en 0, escalada a 1 para el gráfico estándar)
x_t = np.linspace(-4, 4, 100)
plt.plot(x_t, stats.t.pdf(x_t, grados_libertad), label=f'Distribución t ({grados_libertad} df)')

# Región de rechazo (bilateral)
x_rechazo_inferior = np.linspace(-4, valor_critico_t_inferior, 100)
plt.fill_between(x_rechazo_inferior, 0, stats.t.pdf(x_rechazo_inferior, grados_libertad), color='red', alpha=0.5, label=f'Región de Rechazo (α/2 = {nivel_significancia/2})')

x_rechazo_superior = np.linspace(valor_critico_t_superior, 4, 100)
plt.fill_between(x_rechazo_superior, 0, stats.t.pdf(x_rechazo_superior, grados_libertad), color='red', alpha=0.5) # No need for label again

# Valores críticos
plt.axvline(valor_critico_t_inferior, color='red', linestyle='dashed', linewidth=1, label=f'Valores Críticos t ({valor_critico_t_inferior:.2f}, {valor_critico_t_superior:.2f})')
plt.axvline(valor_critico_t_superior, color='red', linestyle='dashed', linewidth=1) # No need for label again

# Estadístico t calculado
plt.axvline(estadistico_t, color='blue', linestyle='dashed', linewidth=1, label=f'Estadístico t Calculado ({estadistico_t:.2f})')

# Configurar el gráfico
plt.title('Prueba de Hipótesis Bilateral (t-test)')
plt.xlabel('Valor t')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()


# 7. Tomar una decisión estadística y concluir
print("7. Tome una decisión estadística y concluya:")
print(f"   Nivel de significancia (α): {nivel_significancia}")
print(f"   Valor p: {valor_p:.4f}")

if valor_p < nivel_significancia:
    decision = "Rechazar H0"
    conclusion = f"Con un nivel de significancia de {nivel_significancia}, hay suficiente evidencia estadística para rechazar la hipótesis nula (H0). Esto sugiere que la duración promedio de los focos ({media_muestral} horas) es significativamente diferente del valor afirmado por el fabricante ({media_poblacional_hipotetizada} horas)."
else:
    decision = "No rechazar H0"
    conclusion = f"Con un nivel de significancia de {nivel_significancia}, no hay suficiente evidencia estadística para rechazar la hipótesis nula (H0). Esto sugiere que la diferencia observada en la media muestral ({media_muestral} horas) respecto a la duración hipotetizada ({media_poblacional_hipotetizada} horas) podría deberse simplemente a la variación aleatoria del muestreo. No podemos concluir que la duración promedio sea significativamente diferente de 10,000 horas basándonos en esta muestra."

print(f"   Decisión: {decision}")
print(f"   Conclusión: {conclusion}")
print("\nExplicación del valor p:")
print(f"   El valor p ({valor_p:.4f}) representa la probabilidad de obtener una media muestral tan extrema o más extrema (en cualquier dirección, ya que es una prueba bilateral) como la observada ({media_muestral} horas), asumiendo que la hipótesis nula (H0) es verdadera (es decir, que la duración promedio real es 10,000 horas).")
print(f"   Dado que el valor p ({valor_p:.4f}) es {'menor' if valor_p < nivel_significancia else 'mayor o igual'} que el nivel de significancia (α = {nivel_significancia}), {'rechazamos' if valor_p < nivel_significancia else 'no rechazamos'} H0.")
```

31. Una compañía de telefonía afirma que la varianza del tiempo de duración de las llamadas es de
σ
2 = 2,5 minutos2
. Se toma una muestra de n = 16 llamadas y se obtiene una varianza muestral
de s
2 = 4. ¿Hay evidencia al 5 % de significancia de que la varianza poblacional difiere de 2.5?


```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Datos proporcionados
varianza_poblacional_hipotetizada = 2.5  # σ²0: Varianza poblacional afirmada en minutos²
n_muestra = 16  # Tamaño de la muestra
varianza_muestral = 4  # s²: Varianza muestral en minutos²
nivel_significancia = 0.05  # α: Nivel de significancia

# 1. Plantear las hipótesis nula y alternativa
# H0: La varianza poblacional es igual a la afirmada (σ² = 2.5 minutos²)
# H1: La varianza poblacional es diferente a la afirmada (σ² ≠ 2.5 minutos²) - Prueba bilateral

print("1. Plantear las hipótesis nula y alternativa:")
print(f"   H0: σ² = {varianza_poblacional_hipotetizada} minutos² (La varianza poblacional es 2.5)")
print(f"   H1: σ² ≠ {varianza_poblacional_hipotetizada} minutos² (La varianza poblacional es diferente a 2.5 - Prueba bilateral)")
print("-" * 30)

# 2. Calcular el estadístico de prueba (Estadístico Chi-cuadrado)
# Fórmula: χ² = ((n - 1) * s²) / σ²0
grados_libertad = n_muestra - 1
estadistico_chi2 = (grados_libertad * varianza_muestral) / varianza_poblacional_hipotetizada

print("2. Calcule el estadístico χ²:")
print(f"   Grados de libertad: {grados_libertad}")
print(f"   Estadístico χ² calculado: {estadistico_chi2:.2f}")
print("-" * 30)

# 3. Determinar los valores críticos y la región de rechazo
# Para una prueba bilateral con α = 0.05 y {grados_libertad} grados de libertad,
# el área en cada cola es α / 2 = 0.025.
# Usamos la función ppf (percent point function) de la distribución chi-cuadrado para encontrar los valores críticos
valor_critico_chi2_inferior = stats.chi2.ppf(nivel_significancia / 2, grados_libertad)
valor_critico_chi2_superior = stats.chi2.ppf(1 - nivel_significancia / 2, grados_libertad)

print("3. Determine los valores críticos y la región de rechazo:")
print(f"   Nivel de significancia (α): {nivel_significancia}")
print(f"   Valores críticos de χ² (prueba bilateral, {grados_libertad} df): {valor_critico_chi2_inferior:.2f} y {valor_critico_chi2_superior:.2f}")
print(f"   Región de rechazo: χ² < {valor_critico_chi2_inferior:.2f} o χ² > {valor_critico_chi2_superior:.2f}")
print("-" * 30)

# 4. Calcular el valor p
# Para una prueba bilateral de chi-cuadrado, el valor p es la suma de las áreas en las colas.
# Si el estadístico χ² calculado es menor que la media de la distribución (que es igual a los grados de libertad),
# el valor p es 2 * el área a la izquierda del estadístico.
# Si el estadístico χ² calculado es mayor que la media, el valor p es 2 * el área a la derecha del estadístico.
# O de forma más general: p-value = 2 * min(P(χ² <= estadistico_chi2), P(χ² >= estadistico_chi2))
valor_p_izquierdo = stats.chi2.cdf(estadistico_chi2, grados_libertad)
valor_p_derecho = 1 - stats.chi2.cdf(estadistico_chi2, grados_libertad)
valor_p = 2 * min(valor_p_izquierdo, valor_p_derecho)


print("5. Calcule el estadístico de prueba y el valor p:") # Re-imprimir estadístico χ² for clarity
print(f"   Estadístico χ² calculado: {estadistico_chi2:.2f}")
print(f"   Valor p: {valor_p:.4f}")
print("-" * 30)


# 6. Graficar la distribución chi-cuadrado, región de rechazo, valores críticos y estadístico calculado
plt.figure(figsize=(10, 6))
# Rango de valores para graficar la distribución chi-cuadrado
# La distribución chi-cuadrado es asimétrica y solo toma valores positivos.
# El rango debe ser adecuado para los grados de libertad.
x_chi2 = np.linspace(0, stats.chi2.ppf(0.999, grados_libertad), 100)
plt.plot(x_chi2, stats.chi2.pdf(x_chi2, grados_libertad), label=f'Distribución χ² ({grados_libertad} df)')

# Región de rechazo (bilateral)
x_rechazo_inferior = np.linspace(0, valor_critico_chi2_inferior, 100)
plt.fill_between(x_rechazo_inferior, 0, stats.chi2.pdf(x_rechazo_inferior, grados_libertad), color='red', alpha=0.5, label=f'Región de Rechazo (α/2 = {nivel_significancia/2})')

x_rechazo_superior = np.linspace(valor_critico_chi2_superior, stats.chi2.ppf(0.999, grados_libertad), 100)
plt.fill_between(x_rechazo_superior, 0, stats.chi2.pdf(x_rechazo_superior, grados_libertad), color='red', alpha=0.5) # No need for label again

# Valores críticos
plt.axvline(valor_critico_chi2_inferior, color='red', linestyle='dashed', linewidth=1, label=f'Valores Críticos χ² ({valor_critico_chi2_inferior:.2f}, {valor_critico_chi2_superior:.2f})')
plt.axvline(valor_critico_chi2_superior, color='red', linestyle='dashed', linewidth=1) # No need for label again

# Estadístico χ² calculado
plt.axvline(estadistico_chi2, color='blue', linestyle='dashed', linewidth=1, label=f'Estadístico χ² Calculado ({estadistico_chi2:.2f})')

# Configurar el gráfico
plt.title('Prueba de Hipótesis Bilateral (Chi-cuadrado para Varianza)')
plt.xlabel('Valor χ²')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()


# 7. Tomar una decisión estadística y concluir
print("7. Tome una decisión estadística y concluya:")
print(f"   Nivel de significancia (α): {nivel_significancia}")
print(f"   Valor p: {valor_p:.4f}")

if valor_p < nivel_significancia:
    decision = "Rechazar H0"
    conclusion = f"Con un nivel de significancia de {nivel_significancia}, hay suficiente evidencia estadística para rechazar la hipótesis nula (H0). Esto sugiere que la varianza poblacional del tiempo de duración de las llamadas es significativamente diferente de {varianza_poblacional_hipotetizada} minutos²."
else:
    decision = "No rechazar H0"
    conclusion = f"Con un nivel de significancia de {nivel_significancia}, no hay suficiente evidencia estadística para rechazar la hipótesis nula (H0). Esto sugiere que la diferencia observada en la varianza muestral ({varianza_muestral} minutos²) respecto a la varianza hipotetizada ({varianza_poblacional_hipotetizada} minutos²) podría deberse simplemente a la variación aleatoria del muestreo. No podemos concluir que la varianza poblacional sea significativamente diferente de 2.5 minutos² basándonos en esta muestra."

print(f"   Decisión: {decision}")
print(f"   Conclusión: {conclusion}")
print("\nExplicación del valor p:")
print(f"   El valor p ({valor_p:.4f}) representa la probabilidad de obtener una varianza muestral tan extrema o más extrema (en cualquier dirección, ya que es una prueba bilateral) como la observada ({varianza_muestral} minutos²), asumiendo que la hipótesis nula (H0) es verdadera (es decir, que la varianza poblacional real es 2.5 minutos²).")
print(f"   Dado que el valor p ({valor_p:.4f}) es {'menor' if valor_p < nivel_significancia else 'mayor o igual'} que el nivel de significancia (α = {nivel_significancia}), {'rechazamos' if valor_p < nivel_significancia else 'no rechazamos'} H0.")
```

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Datos proporcionados
varianza_poblacional_hipotetizada = 1.2  # σ²0: Varianza poblacional afirmada en ml²
n_muestra = 20  # Tamaño de la muestra
varianza_muestral = 2.0  # s²: Varianza muestral en ml²
nivel_significancia = 0.01  # α: Nivel de significancia

# 1. Plantear las hipótesis nula y alternativa
# H0: La varianza del llenado es igual a la afirmada (σ² = 1.2 ml²)
# H1: La varianza del llenado es mayor a la afirmada (σ² > 1.2 ml²) - Prueba unilateral derecha

print("1. Plantear las hipótesis nula y alternativa:")
print(f"   H0: σ² = {varianza_poblacional_hipotetizada} ml² (La varianza poblacional es 1.2)")
print(f"   H1: σ² > {varianza_poblacional_hipotetizada} ml² (La varianza poblacional es mayor a 1.2 - Prueba unilateral derecha)")
print("-" * 30)

# 2. Calcular el estadístico de prueba (Estadístico Chi-cuadrado)
# Fórmula: χ² = ((n - 1) * s²) / σ²0
grados_libertad = n_muestra - 1
estadistico_chi2 = (grados_libertad * varianza_muestral) / varianza_poblacional_hipotetizada

print("2. Calcule el estadístico χ²:")
print(f"   Grados de libertad: {grados_libertad}")
print(f"   Estadístico χ² calculado: {estadistico_chi2:.2f}")
print("-" * 30)

# 3. Determinar el valor crítico y la región de rechazo
# Para una prueba unilateral derecha con α = 0.01 y {grados_libertad} grados de libertad,
# el valor crítico de χ² es el valor tal que el área a su derecha en la distribución chi-cuadrado es α.
valor_critico_chi2 = stats.chi2.ppf(1 - nivel_significancia, grados_libertad)

print("3. Determine el valor crítico y la región de rechazo:")
print(f"   Nivel de significancia (α): {nivel_significancia}")
print(f"   Valor crítico de χ² (prueba unilateral derecha, {grados_libertad} df): {valor_critico_chi2:.2f}")
print(f"   Región de rechazo: χ² > {valor_critico_chi2:.2f}")
print("-" * 30)

# 4. Calcular el valor p
# Para una prueba unilateral derecha de chi-cuadrado, el valor p es el área a la derecha del estadístico χ² calculado
valor_p = 1 - stats.chi2.cdf(estadistico_chi2, grados_libertad)

print("5. Calcule el estadístico de prueba y el valor p:") # Re-imprimir estadístico χ² for clarity
print(f"   Estadístico χ² calculado: {estadistico_chi2:.2f}")
print(f"   Valor p: {valor_p:.4f}")
print("-" * 30)


# 6. Graficar la distribución chi-cuadrado, región de rechazo, valor crítico y estadístico calculado
plt.figure(figsize=(10, 6))
# Rango de valores para graficar la distribución chi-cuadrado
# La distribución chi-cuadrado es asimétrica y solo toma valores positivos.
# El rango debe ser adecuado para los grados de libertad y el estadístico calculado.
x_chi2 = np.linspace(0, max(estadistico_chi2 + 5, stats.chi2.ppf(0.999, grados_libertad)), 100) # Adjust max x for better visualization
plt.plot(x_chi2, stats.chi2.pdf(x_chi2, grados_libertad), label=f'Distribución χ² ({grados_libertad} df)')

# Región de rechazo (unilateral derecha)
x_rechazo = np.linspace(valor_critico_chi2, max(estadistico_chi2 + 5, stats.chi2.ppf(0.999, grados_libertad)), 100) # Adjust max x for better visualization
plt.fill_between(x_rechazo, 0, stats.chi2.pdf(x_rechazo, grados_libertad), color='red', alpha=0.5, label=f'Región de Rechazo (α = {nivel_significancia})')

# Valor crítico
plt.axvline(valor_critico_chi2, color='red', linestyle='dashed', linewidth=1, label=f'Valor Crítico χ² ({valor_critico_chi2:.2f})')

# Estadístico χ² calculado
plt.axvline(estadistico_chi2, color='blue', linestyle='dashed', linewidth=1, label=f'Estadístico χ² Calculado ({estadistico_chi2:.2f})')

# Configurar el gráfico
plt.title('Prueba de Hipótesis Unilateral Derecha (Chi-cuadrado para Varianza)')
plt.xlabel('Valor χ²')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()


# 7. Tomar una decisión estadística y concluir
print("7. Tome una decisión estadística y concluya:")
print(f"   Nivel de significancia (α): {nivel_significancia}")
print(f"   Valor p: {valor_p:.4f}")

if valor_p < nivel_significancia:
    decision = "Rechazar H0"
    conclusion = f"Con un nivel de significancia de {nivel_significancia}, hay suficiente evidencia estadística para rechazar la hipótesis nula (H0). Esto sugiere que la varianza poblacional del llenado de las botellas es significativamente mayor a {varianza_poblacional_hipotetizada} ml²."
else:
    decision = "No rechazar H0"
    conclusion = f"Con un nivel de significancia de {nivel_significancia}, no hay suficiente evidencia estadística para rechazar la hipótesis nula (H0). Esto sugiere que la diferencia observada en la varianza muestral ({varianza_muestral} ml²) respecto a la varianza hipotetizada ({varianza_poblacional_hipotetizada} ml²) podría deberse simplemente a la variación aleatoria del muestreo. No podemos concluir que la varianza poblacional sea significativamente mayor a 1.2 ml² basándonos en esta muestra."

print(f"   Decisión: {decision}")
print(f"   Conclusión: {conclusion}")
print("\nExplicación del valor p:")
print(f"   El valor p ({valor_p:.4f}) representa la probabilidad de obtener una varianza muestral tan extrema o más extrema ({varianza_muestral} ml²) que la observada, asumiendo que la hipótesis nula (H0) es verdadera (es decir, que la varianza poblacional real es 1.2 ml²).")
print(f"   Dado que el valor p ({valor_p:.4f}) es {'menor' if valor_p < nivel_significancia else 'mayor o igual'} que el nivel de significancia (α = {nivel_significancia}), {'rechazamos' if valor_p < nivel_significancia else 'no rechazamos'} H0.")
```

concluir al 5 % que la varianza poblacional es menor que la garantizada?

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Datos proporcionados
varianza_poblacional_hipotetizada = 0.5  # σ²0: Varianza poblacional garantizada (mínima) en mm²
n_muestra = 12  # Tamaño de la muestra
varianza_muestral = 0.3  # s²: Varianza muestral en mm²
nivel_significancia = 0.05  # α: Nivel de significancia

# 1. Plantear las hipótesis nula y alternativa
# H0: La varianza poblacional es igual al mínimo garantizado (σ² = 0.5 mm²)
# H1: La varianza poblacional es menor que el mínimo garantizado (σ² < 0.5 mm²) - Prueba unilateral izquierda

print("1. Plantear las hipótesis nula y alternativa:")
print(f"   H0: σ² = {varianza_poblacional_hipotetizada} mm² (La varianza poblacional es 0.5)")
print(f"   H1: σ² < {varianza_poblacional_hipotetizada} mm² (La varianza poblacional es menor a 0.5 - Prueba unilateral izquierda)")
print("-" * 30)

# 2. Calcular el estadístico de prueba (Estadístico Chi-cuadrado)
# Fórmula: χ² = ((n - 1) * s²) / σ²0
grados_libertad = n_muestra - 1
estadistico_chi2 = (grados_libertad * varianza_muestral) / varianza_poblacional_hipotetizada

print("2. Calcule el estadístico χ²:")
print(f"   Grados de libertad: {grados_libertad}")
print(f"   Estadístico χ² calculado: {estadistico_chi2:.2f}")
print("-" * 30)

# 3. Determinar el valor crítico y la región de rechazo
# Para una prueba unilateral izquierda con α = 0.05 y {grados_libertad} grados de libertad,
# el valor crítico de χ² es el valor tal que el área a su izquierda en la distribución chi-cuadrado es α.
valor_critico_chi2 = stats.chi2.ppf(nivel_significancia, grados_libertad)

print("3. Determine el valor crítico y la región de rechazo:")
print(f"   Nivel de significancia (α): {nivel_significancia}")
print(f"   Valor crítico de χ² (prueba unilateral izquierda, {grados_libertad} df): {valor_critico_chi2:.2f}")
print(f"   Región de rechazo: χ² < {valor_critico_chi2:.2f}")
print("-" * 30)

# 4. Calcular el valor p
# Para una prueba unilateral izquierda de chi-cuadrado, el valor p es el área a la izquierda del estadístico χ² calculado
valor_p = stats.chi2.cdf(estadistico_chi2, grados_libertad)

print("5. Calcule el estadístico de prueba y el valor p:") # Re-imprimir estadístico χ² for clarity
print(f"   Estadístico χ² calculado: {estadistico_chi2:.2f}")
print(f"   Valor p: {valor_p:.4f}")
print("-" * 30)


# 6. Graficar la distribución chi-cuadrado, región de rechazo, valor crítico y estadístico calculado
plt.figure(figsize=(10, 6))
# Rango de valores para graficar la distribución chi-cuadrado
# La distribución chi-cuadrado es asimétrica y solo toma valores positivos.
# El rango debe ser adecuado para los grados de libertad y el estadístico calculado.
x_chi2 = np.linspace(0, stats.chi2.ppf(0.99, grados_libertad), 100) # Adjust max x for better visualization
plt.plot(x_chi2, stats.chi2.pdf(x_chi2, grados_libertad), label=f'Distribución χ² ({grados_libertad} df)')

# Región de rechazo (unilateral izquierda)
x_rechazo = np.linspace(0, valor_critico_chi2, 100)
plt.fill_between(x_rechazo, 0, stats.chi2.pdf(x_rechazo, grados_libertad), color='red', alpha=0.5, label=f'Región de Rechazo (α = {nivel_significancia})')

# Valor crítico
plt.axvline(valor_critico_chi2, color='red', linestyle='dashed', linewidth=1, label=f'Valor Crítico χ² ({valor_critico_chi2:.2f})')

# Estadístico χ² calculado
plt.axvline(estadistico_chi2, color='blue', linestyle='dashed', linewidth=1, label=f'Estadístico χ² Calculado ({estadistico_chi2:.2f})')

# Configurar el gráfico
plt.title('Prueba de Hipótesis Unilateral Izquierda (Chi-cuadrado para Varianza)')
plt.xlabel('Valor χ²')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()


# 7. Tomar una decisión estadística y concluir
print("7. Tome una decisión estadística y concluya:")
print(f"   Nivel de significancia (α): {nivel_significancia}")
print(f"   Valor p: {valor_p:.4f}")

if valor_p < nivel_significancia:
    decision = "Rechazar H0"
    conclusion = f"Con un nivel de significancia de {nivel_significancia}, hay suficiente evidencia estadística para rechazar la hipótesis nula (H0). Esto sugiere que la varianza poblacional en el grosor de las piezas metálicas es significativamente menor que el mínimo garantizado de {varianza_poblacional_hipotetizada} mm²."
else:
    decision = "No rechazar H0"
    conclusion = f"Con un nivel de significancia de {nivel_significancia}, no hay suficiente evidencia estadística para rechazar la hipótesis nula (H0). Esto sugiere que la diferencia observada en la varianza muestral ({varianza_muestral} mm²) respecto a la varianza hipotetizada ({varianza_poblacional_hipotetizada} mm²) podría deberse simplemente a la variación aleatoria del muestreo. No podemos concluir que la varianza poblacional sea significativamente menor que 0.5 mm² basándonos en esta muestra."

print(f"   Decisión: {decision}")
print(f"   Conclusión: {conclusion}")
print("\nExplicación del valor p:")
print(f"   El valor p ({valor_p:.4f}) representa la probabilidad de obtener una varianza muestral tan extrema o más extrema ({varianza_muestral} mm²) que la observada, asumiendo que la hipótesis nula (H0) es verdadera (es decir, que la varianza poblacional real es 0.5 mm²).")
print(f"   Dado que el valor p ({valor_p:.4f}) es {'menor' if valor_p < nivel_significancia else 'mayor o igual'} que el nivel de significancia (α = {nivel_significancia}), {'rechazamos' if valor_p < nivel_significancia else 'no rechazamos'} H0.")
```
